```python
items = list(range(0, 250, 20))
rel_cols = [
        'name',
        'distanceFromTarryallCabin',
        'avgRating',
        'reviewsCount',
        'isNewListing',
        'isSuperhost',
        'previewAmenityNames',
        # 'primaryLine.accessibilityLabel',
        'title',
        'personCapacity',
        'structuredContent.primaryLine',      
    ]
```

```python

# set up all the requests to the api for 20-items-per-page approach
final_res = list()
for item in items:
    querystring['variables'] = variables_builder(item)
    final_res.append(
        requests.request("GET", url, headers=headers, params=querystring)
    )
    
    
#  drilldown to relevant section in airbnb (as of Aug2022)
json_total_data = list(
    map(
        lambda response:
        response.json()['data']['presentation']['explore']['sections']['sections'][2]['section']['child']['section']['items'],
        final_res
    )
)

#  some vals are getting mixed, store to QA
backup_json_total = json_total_data


#  update/consolidate information into a single dict
chain(
    *list(
        map(
            lambda stays_chunk:
                map(
                lambda stay:
                    stay['listing'].update(stay['pricingQuote']['structuredStayDisplayPrice']),
                    stays_chunk
            ),
            json_total_data
        )
    )
)

#  merge list of lists for ease of use
json_total_listings = list(
    chain(
        *json_total_data
    )
)


#  create the dataframe - it's massive column-wise
listings_detail = pd.concat(
    list(
        map(
            lambda stay:
                pd.json_normalize(
                    stay['listing']
                ),
            json_total_listings
        )
    )
)

#  pull bedding content out of content
listings_detail.loc[:, 'structuredContent.primaryLine'] = listings_detail.loc[:, 'structuredContent.primaryLine'].apply(
    lambda x: ';'.join([m['body'] for m in x])
)

#  redundant but im lazy - rename for my notebooks sake
all_listings_in_jefferson = listings_detail
```

```python
#  merge lat, long into tuple
all_listings_in_jefferson.loc[:, 'coords'] = all_listings_in_jefferson.apply(
    lambda x: (x.lat, x.lng),
    axis=1
)
```

```python
import geopy.distance

#  from above - use as static metric in below calc
tarryall = (39.296114, -105.7186119)

#  calculate distance from cabin, in miles
all_listings_in_jefferson.loc[:, 'distanceFromTarryallCabin'] = all_listings_in_jefferson.coords.apply(
    lambda coordinates:
    geopy.distance.geodesic(tarryall, coordinates).mi
)
#  filter down and apply sort based on distance from cabin
all_listings_in_jefferson = listings_detail[rel_cols].sort_values('distanceFromTarryallCabin')

all_listings_in_jefferson.loc[:, 'distanceFromTarryallCabin'] = all_listings_in_jefferson.distanceFromTarryallCabin.apply(
    lambda coordinates:
    f'{coordinates} miles'
)
```
